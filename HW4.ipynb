#1. LeetCode 225. Implement Stack using Queues
* 큐만 사용해서 스택을 구현하는 문제
* 스택은 후입선출, 큐는 선입선출 구조
* 'push(), pop(), top(), empty()'함수 구현

```python
from collections import deque

class MyStack:
  def __init__(self):
    self.q=deque()

    def push(self, x:int)->None:
      self.q.append(x)
      for _ in range(len(self.q)-1):
        self.q.append(self.q.popleft())

        def pop(self)->int:
          return self.q.popleft()

          def top(self)->int:
            return self.q[0]

            def empty(self)->bool:
              return not self.q
```

풀이 설명
push(x): 새 원소를 큐에 추가한 후, 앞의 원소들을 뒤로 이동
pop() :큐의 맨 앞 원소를 제거
top() : 큐의 맨 앞 원소를 반환
empty() : 큐가 비어있지 확인

**시간 복잡도
push(): O(N)
pop(), top(), empty() : O(1)
